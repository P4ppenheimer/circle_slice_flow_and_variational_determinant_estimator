```python
import pandas as pd 
import matplotlib.pyplot as plt
```

### 3D fit

```python
file_name = '3D_fit.csv'

df0 = pd.read_csv(file_name)
df0.tail()
```

```python
df0.columns.tolist()
```

```python
col_list = ['cou_m_NC_8_NF_3 - KL_test','cou_s_NB_8_NF_3 - KL_test','ar_cyl_m_NC_8_NF_3 - KL_test','ar_cyl_s_NB_8_NF_3 - KL_test']

df = df0[col_list]
```

```python
df.head(10)
```

```python
df.idxmin()
```

```python
df.iloc[-1].round(3)
```

```python
df.min().round(3)
```

```python
col_list = ['cou_m_NC_8_NF_3 - KL_test','cou_s_NB_8_NF_3 - KL_test','ar_cyl_m_NC_8_NF_3 - KL_test','ar_cyl_s_NB_8_NF_3 - KL_test']

renamed_col_list = [ el.strip(' - KL_test') for el in col_list]

df.columns = renamed_col_list 
```

```python
plt.figure(dpi=200)

ax = df.plot(grid=True, ax = plt.gca(), xlabel='Evaluation Steps', ylabel='KL test', ylim=(0,0.2))
ax.legend(['Circle Slice Moebius','Circle Slice Spline','AR Cylindrical Moebius','AR Cylindrical Spline'])
file_name = '3D_fit.svg'

plt.savefig(file_name, format='svg')
```

### NLL plots with different seeds

#### d=32

```python
file_name = 'dim_32_nll_all_seeds.csv'

df0 = pd.read_csv(file_name)
df0.tail()
```

```python
# df0.columns.tolist()
```

```python
col_list = ['cou_m_NC_8_NF_3_seed_07 - neg_log_likel_test',
         'cou_cyl_m_NC_8_NF_3_seed_07 - neg_log_likel_test',
         'cou_m_NC_8_NF_3_seed_24 - neg_log_likel_test',
         'cou_cyl_m_NC_8_NF_3_seed_24 - neg_log_likel_test']

        
         

df = df0[col_list]

renamed_col_list = [ el.strip(' - neg_log_likel_test') for el in col_list]

df.columns = renamed_col_list 
df.head()
```

```python
# plt.figure(dpi=200)
# df.plot(grid=True, ax = plt.gca(), xlabel='Evaluation Steps', ylabel='KL test')
```

```python
plt.figure(dpi=200)
styles = ['-','-','--','--']
ax = df.plot(grid=True, ax = plt.gca(), xlabel='Evaluation Steps', ylabel='nll test',style = styles)

ax.legend(['Circle Slice Moebius Run B', 'Cylindrical Moebius Run B','Circle Slice Moebius Run C', 'Cylindrical Moebius Run C'])

file_name = '32D_fit_nll_mult_seeds.svg'

plt.savefig(file_name, format='svg')
```
